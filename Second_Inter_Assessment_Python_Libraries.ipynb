```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Read the data

```python
filepath='/content/adult.csv'
df_adult=pd.read_csv(filepath)
df_adult
```

Display the first 5 rows of the dataset.

summary of the DataFrame

```python
df_adult.info()
```

descriptive statistics for numeric columns

```python
df_adult.describe()
```

Display the column names of the DataFrame.

```python
display(df_adult.columns)
```

unique values in the 'workclass' column.

```python
display(df_adult['workclass'].unique())
```

Create a new column 'income_binary' where <=50K is 0 and >50K is 1.

```python
df_adult['income_binary'] = df_adult['income'].apply(lambda x: 0 if x == '<=50K' else 1)
display(df_adult[['income', 'income_binary']].head())
```

Create a new column 'age_group' categorizing ages into 'Youth' (< 30), 'Adult' (30-60), and 'Senior' (60+).

```python
bins = [-np.inf, 30, 60, np.inf]
labels = ['Youth', 'Adult', 'Senior']
df_adult['age_group'] = pd.cut(df_adult['age'], bins=bins, labels=labels, right=False)
display(df_adult[['age', 'age_group']].head())
```

Rename the 'educational-num' column to 'education_num'.

```python
df_adult = df_adult.rename(columns={'educational-num': 'education_num'})
display(df_adult.head())
```

Replace the missing values represented by '?' in 'workclass' and 'occupation' columns with 'Unknown'.

```python
df_adult['workclass'] = df_adult['workclass'].replace('?', 'Unknown')
df_adult['occupation'] = df_adult['occupation'].replace('?', 'Unknown')
display(df_adult[['workclass', 'occupation']].head())
```

Calculate the average age for each workclass.

```python
average_age_by_workclass = df_adult.groupby('workclass')['age'].mean()
display(average_age_by_workclass)
```

Find the total capital-gain for each gender.

```python
total_capital_gain_by_gender = df_adult.groupby('gender')['capital-gain'].sum()
display(total_capital_gain_by_gender)
```

Get the count of each marital-status category.

```python
marital_status_counts = df_adult['marital-status'].value_counts()
display(marital_status_counts)
```

Find the mean hours-per-week worked by individuals in different education levels.

```python
mean_hours_by_education = df_adult.groupby('education')['hours-per-week'].mean()
display(mean_hours_by_education)
```

Calculate the mean, median, and standard deviation of 'fnlwgt' for each race.

```python
fnlwgt_stats_by_race = df_adult.groupby('race')['fnlwgt'].agg(['mean', 'median', 'std'])
display(fnlwgt_stats_by_race)
```

Get the unique values and their counts for the 'native-country' column

```python
native_country_counts = df_adult['native-country'].value_counts()
display(native_country_counts)
```

Filter the dataset to include only individuals with income >50K.

```python
df_income_greater_50K = df_adult[df_adult['income'] == '>50K']
display(df_income_greater_50K.head())
```

Sort the dataset by age in descending order.

```python
df_adult_sorted_by_age = df_adult.sort_values(by='age', ascending=False)
display(df_adult_sorted_by_age.head())
```

Filter the dataset to include only individuals who work more than 40 hours per week.

```python
df_more_than_40_hours = df_adult[df_adult['hours-per-week'] > 40]
display(df_more_than_40_hours.head())
```

Calculate the percentage of missing values in each column.

```python
missing_values_percentage = (df_adult.isnull().sum() / len(df_adult)) * 100
display(missing_values_percentage)
```

Calculate the median income by education level and gender.

```python
median_income_by_education_gender = df_adult.groupby(['education', 'gender'])['income_binary'].median()
print(median_income_by_education_gender)
```

Find the top 3 occupations with the highest average capital-gain.

```python
top_3_occupations_capital_gain = df_adult.groupby('occupation')['capital-gain'].mean().sort_values(ascending=False).head(3)
display(top_3_occupations_capital_gain)
```

Filter the dataset to include individuals who are 'Married-civ-spouse' and have an income >50K.

```python
df_married_high_income = df_adult[(df_adult['marital-status'] == 'Married-civ-spouse') & (df_adult['income'] == '>50K')]
display(df_married_high_income.head())
```

Create 5 different charts and write your observations from the chart as comment

```python
plt.figure(figsize=(10, 8))
marital_status_counts = df_adult['marital-status'].value_counts()
plt.pie(marital_status_counts, labels=marital_status_counts.index, autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.title('Distribution of Marital Status')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='education_num', y='income_binary',data=df_adult, palette='magma')
plt.title('Proportion of >50K Income by Education Level (Numerical)')
plt.xlabel('Education Level (Numerical)')
plt.ylabel('Proportion of Income >50K')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='gender', y='income_binary', data=df_adult, palette='coolwarm')
plt.title('Proportion of >50K Income by Gender')
plt.xlabel('Gender')
plt.ylabel('Proportion of Income >50K')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.barplot(x='workclass', y='income_binary', data=df_adult, palette='viridis')
plt.title('Proportion of >50K Income by Workclass')
plt.xlabel('Workclass')
plt.ylabel('Proportion of Income >50K')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_adult['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
