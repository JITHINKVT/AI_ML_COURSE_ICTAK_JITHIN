```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/tip-amount.xlsx')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df[['Observed total bill(xi)','Observed tip amount(yi)']].corr()
```

```python
sns.scatterplot(x='Observed total bill(xi)',y='Observed tip amount(yi)',data=df)
plt.title("Bill vs Tip")
plt.show()
```

```python
#y = mx+b

x = df['Observed total bill(xi)'] #x1,x2,x3
y = df['Observed tip amount(yi)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.values.reshape(-1, 1), y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
y_pred=model.predict(x_test.values.reshape(-1, 1))
```

```python
plt.scatter(x_test, y_test, color='blue',label='Actual')
plt.plot(x_test,y_pred,color='red',label='Predicted')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y_test,y_pred)))
print("R2:",r2_score(y_test,y_pred))
```
